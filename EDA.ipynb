```python
xls = pd.ExcelFile(path)
df = pd.read_excel(xls, sheet_name=sheet_names[0], skiprows=8)

for sheet in sheet_names[1:]:
    df = pd.concat([df, pd.read_excel(xls, sheet_name=sheet, skiprows=8)])
    

df.tail()
```

```python
df.isnull().sum()
```

```python
df["Rs (W/m²)"].fillna(df["RS(W/m²)"]).isnull().sum()
```

```python
df["Rs (W/m²)"] = df["Rs (W/m²)"].fillna(df["RS(W/m²)"])

df.info()
```

```python
del df["RS(W/m²)"], df["Foyo (2017)"], df["kt"], df["Rs0"], df["JD"]
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.rename(columns={"AAAA/MM/DD HH:MM HORAZ": "Timestamp", "T Aire (°C)": "T Aire (C)"}, inplace=True)

df.columns
```

```python
df.tail()
```

```python
df.reset_index(drop=True, inplace=True)
df.tail()
```

```python
pd.to_datetime(df["Timestamp"], errors="coerce").isnull().sum()
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")

df.head(2)
```

```python
df.info()
```

```python
df["Year"] = df["Timestamp"].dt.year
df["Mes"] = df["Timestamp"].dt.month
df["Dia"] = df["Timestamp"].dt.day

df.head()
```

```python
df.tail()
```

```python
column_order = ["Timestamp", "Year", "Mes", "Dia"] + [col for col in df.columns if col not in ["Timestamp", "Year", "Mes", "Dia"]]
# column_order = ["Timestamp", "Year", "Mes", "Dia", ]
df = df[column_order]
df.head()
```

```python
del df["Beta"]
df.info()
```

```python
df.to_csv("Chihuahua_clima.csv")
df.to_json("Chihuahua_clima.json")
df.to_excel("Chihuahua_clima.xlsx")
```
