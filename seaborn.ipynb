```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

np.random.seed(614)
```

```python
def create_trigon():
    x = np.linspace(0, 10, 100)
    #
    #
    return (x, np.sin(x), np.cos(x))

def random_line():
    #
    return np.linspace(0, 2, 15)

def exp_line():
    return random_line() ** 2


def log_line():
    return np.log(random_line())
```

```python
x, seno, coseno = create_trigon()
```

```python
plt.title("Seaborn lineplot")
sns.lineplot([seno, coseno])
plt.grid(True)
plt.show()
```

```python
seno.shape
```

```python
line, exp, log = random_line(), exp_line(), log_line()


plt.plot(line, line, "-o", label="Lineal")
plt.plot(line, exp, "-^", label="Exponencial")
plt.plot(line, log, "-x", label="Logaritmica")

plt.legend(loc="best")
plt.title("Comparacin de lneas")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
```

```python
mask = np.random.randn(100) < 0 #
genres = np.empty_like(mask, dtype=object)
genres[mask] = "Female"
genres[np.bitwise_not(mask)] = "Male"
genres
```

```python
qualified = np.random.randn(100) < 0
qualified[qualified != 0] = 1
qualified
```

```python
count = pd.DataFrame({"genres": genres, "qualified": qualified})
sns.countplot(count, x="genres", hue="qualified")
plt.title("Count Plot")
plt.show()
```
